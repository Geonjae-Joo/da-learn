```python
import pandas as pd
import numpy as np

import warnings
warnings.filterwarnings('ignore')
```

pd.merge( on =키이름, how='outer' or 'inner'): 하나 이상의 키를 따라 행을 이어 붙인다.  
pd.concat : 하나 축을 따라 객체를 이어 붙인다.

merge 와 join 은 관계형 데이터베이스의 핵심 연산  
키(key)를 사용하여 로우를 합친다

```python
df1 = pd.DataFrame({'key':list('bbacaab'),'key2':list('ddefgge'),'data1':range(7)})
df2 = pd.DataFrame({'key':list('abc'),'key2':list('ggf'),'data2':range(3)})
df1
```

```python
df2
```

pd.merge 공통되는 열의 이름을 기준으로 그 열 안의 공통의 데이터를 추출하여 합친다.  
 -내부에 on 옵션을 걸어 어떤 ㄴ열에 대해 병합할 것인지 설정할 수 있다.
  
  옵션들:
   - on: 어떤 열을 key로 잡을 것인다
   - how: 병합방식  
       inner  
       outer  
       lef  
       right<br>  


현재 df1, df2의 공토외는 열은 'key'와 'key2'이며 , 그 안의 공통 데이터는 'a','g'와 'c','f', 순서쌍이다

```python
# dufault= inner
pd.merge(df1,df2)
```

```python
pd.merge(df1,df2,how='outer')
```

```python

pd.merge(df1,df2,on='key')
```

```python
pd.merge(df1,df2,on='key2',how='outer')
```

```python
pd.merge(df1,df2,on='key',suffixes=['_df1','_df2'])
```

left_on, right_on 옵션  
작동은 merge 와 같다. 즉 left_on과 right_on에 대응되는 값이 key가 되어 병합

```python
df3= pd.DataFrame({'rkey': list('bbacaab'),'data1':range(7)})
df4=pd.DataFrame({'rkey': list('abd'),'data2':range(3)})
df3
```

```python
df4
```

```python
pd.merge(df4,df3,how='left')
```

```python
df3= pd.DataFrame({'lkey': list('bbacaab'),'data1':range(7)})
df3
```

```python
pd.merge(df3,df4,left_on='lkey',right_on='rkey')
```

```python
display(df3)
display(df4)
display(pd.merge(df3,df4,left_on='lkey',right_on='rkey'))
pd.merge(df4,df3,left_on='rkey',right_on='lkey')
```

join의 개념
    inner join : 교집합인 데이터들을 중심으로  
    outer join : 합집합인 데이터들을 중심으로  
    left join : 왼쪽 로우를 포함하는 결과  
    right join : 오른쪽 로우를 포함하는 결과

```python
df1
```

```python
pd.merge(df2,df1,how='left')
```
